```python
import pandas as pd

# Load the dataset (semicolon-separated)
df = pd.read_csv('dataset_Facebook.csv', sep=';')

# a. Create Data Subsets
# -----------------------

# Subset with selected columns
subset_columns = df[['Type', 'like', 'comment', 'share']]

# Subset with only 'Photo' type posts
photo_only = df[df['Type'] == 'Photo']


# b. Merge Data
# --------------

# Simulated external sentiment dataset
sentiment_df = pd.DataFrame({
    'Post ID': df.index[:5],
    'Sentiment': ['Positive', 'Neutral', 'Positive', 'Negative', 'Neutral']
})

# Add 'Post ID' to original dataset
df['Post ID'] = df.index

# Merge on 'Post ID'
merged_df = pd.merge(df, sentiment_df, on='Post ID', how='left')


# c. Sort Data
# -------------

# Sort by 'like' in descending order
sorted_df = df.sort_values(by='like', ascending=False)


# d. Transpose Data
# ------------------

# Transpose the first 3 rows
transposed_df = df.head(3).T


# e. Shape and Reshape Data
# ---------------------------

# Check original shape
original_shape = df.shape

# Melt (long format)
melted_df = pd.melt(df,
                    id_vars=['Type'],
                    value_vars=['like', 'comment', 'share'],
                    var_name='Metric',
                    value_name='Value')

# Pivot (wide format with mean)
pivoted_df = melted_df.pivot_table(index='Type',
                                    columns='Metric',
                                    values='Value',
                                    aggfunc='mean')

# Show results
print("Subset Columns (Type, Like, Comment, Share):\n", subset_columns.head())
print("\nSubset Only 'Photo' Type:\n", photo_only.head())
print("\nMerged Data with Sentiment:\n", merged_df[['Post ID', 'Type', 'Sentiment']].head())
print("\nSorted by 'Like' Descending:\n", sorted_df[['Type', 'like']].head())
print("\nTransposed First 3 Rows:\n", transposed_df)
print("\nOriginal Shape:", original_shape)
print("\nMelted Data (Long Format):\n", melted_df.head())
print("\nPivoted Data (Mean Interactions by Type):\n", pivoted_df)
```
